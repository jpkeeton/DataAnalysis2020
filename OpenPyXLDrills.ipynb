OpenPyXL



```python
# create a simple spreadsheet
from openpyxl import Workbook
```

```python
# create a workbook
workbook = Workbook()
# reference a sheet
sheet = workbook.active
# add some strings to the sheets
sheet['A1'] ='Hello'
sheet['B2'] ='Excel'
```

```python
# Save the file
workbook.save(filename='hello_Excel.xlsx')
```
